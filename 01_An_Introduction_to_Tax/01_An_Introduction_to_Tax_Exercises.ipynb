**Chuck, a single taxpayer, earns 75,000 in taxable income and 10,000 in interest from an investment in City of Heflin bonds.**

1. How much federal tax will he owe? 
2. What is his average tax rate? 
3. What is his effective tax rate? 
4. What is his current marginal tax rate?

```python
status = "SNG" # Single
taxable_income = 75000 
interest_income = 10000
change_in_income = 0
```

```python
tax = Tax(status, taxable_income)
rate = Planning(status, taxable_income, interest_income, change_in_income)
print(f"1. He will owe: ${tax.compute_tax():,.0f}") 
print(f"2. His Average Tax Rate is: {rate.average_tax_rate()*100:.2f}%")
print(f"3. His Effective Tax Rate is: {rate.effective_tax_rate()*100:.2f}%")
print(f"2. His Current Marginal Tax Rate (per 2026 tables) is: {22:.2f}%")
```

**Using the facts in the previous problem:** 

1. If Chuck earns an additional 40,000 of taxable income, what is his marginal tax rate on this income? 
2. What is his marginal rate if, instead, he had 40,000 of additional deductions?

```python
change_in_income = {
    "1": ["Additional Income", 40000], 
    "2": ["Additional Deduction", -40000],
}

for change in change_in_income.keys():
    marginal = Planning(status, taxable_income, interest_income, change_in_income[change][1])
    print(f"{change}. Marginal Tax Rate of {change_in_income[change][0]} of ${change_in_income[change][1]:,.0f} = {marginal.marginal_tax_rate()*100:.2f}%")
```
