**Example_1-3 - Income Tax** 

Bill and Mercedes, file a joint tax return. They have 160,000 of taxable income this year (gross income minus deductions). Assuming the following federal tax rate schedule applies, how much federal income tax will they owe this year?

```python
taxable_income = 160000
classification = "MFJ"
federal_income_tax = Tax(classification, taxable_income).compute_tax()
print(f"They will owe ${federal_income_tax:,.2f} in federal income taxes this year.")
```

**Example_1-4 - Marginal Tax Rates**

Bill is a well-known economics professor, he signs a publishing contract with an 80,000 royalty advance. What would Bill and Mercedes’s marginal tax rate be on this 80,000 of additional income?


```python
additional_income = 80000
tax = Planning(classification, taxable_income, additional_income)
print(f"Marginal tax rate on additional income: {tax.marginal_tax_rate()*100:.2f}%")
```

**Example 1-5 - Marginal Tax Rates**

Assume now that, instead of receiving a book advance, Bill and Mercedes start a new business that loses 90,000 this year (it results in 90,000 of additional deductions). What would be their marginal tax rate for these additional deductions?


```python
additional_loss = -90000
tax = Planning(classification, taxable_income, additional_loss)
print(f"Marginal tax rate on additional deduction: {tax.marginal_tax_rate()*100:.2f}%")
```

**Example 1-6 - Average Tax Rates**

Assuming Bill and Mercedes have 160,000 of taxable income and 10,000 of nontaxable income, what is their average tax rate?

```python
non_taxable_income = 10000 
rate = Planning(classification, taxable_income, non_taxable_income=non_taxable_income)
average_tax_rate = rate.average_tax_rate()
print(f"Average Tax Rate is: {average_tax_rate*100:,.2f}%")
print("Note: Nontaxable income given but irrelevant to calculation.")
```

**Example 1-7 - Effective Tax Rates**

Again, given the same income figures as in `Example 1-6` (160,000 of taxable income and 10,000 of nontaxable income), what is Bill and Mercedes’s effective tax rate?

```python
effective_tax_rate = rate.effective_tax_rate()
print(f"Effective tax rate: {effective_tax_rate*100:.2f}%")
print("Note: In this case, nontaxable income is used in the calculation")
```

**Example 1-8 - Proportional Tax Rates**

Knowing her dad is a serious Georgia Bulldog fan, Margaret buys a 100 sweatshirt in downtown Athens. The city of Athens imposes a sales tax rate of 7 percent. How much tax does Margaret pay on the purchase?

```python
sweatshirt = 100
tax_rate = .07
print(f"Tax on purchase ${sweatshirt*tax_rate:,.2f}")
```
